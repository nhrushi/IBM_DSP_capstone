### Introduction
Trains are among the best transportation options across Europe. Train networks are responsible for guiding 8 billion passengers to their destinations across Europe in the year 2018 and this number has been growing ever since. With a large exposure of consumers commuting through train stations, the commercial areas around train stations have been attracting infrastructure investments.
In this project, we will analyze the most common investments near major train stations across Europe and trends among services provided across various regions in Europe.

### Business problem

This project considers data related to locations of train stations in major European cities and analyzes the existing trends related to facilities and services provided to customers close to these train stations.
Main aim of this project is to suggest a new business investment opportunity for establishing franchises, expanding to new markets or create a new service industries near train stations. During this project, we will explore and examine venues (such as cafes, supermarkets, restaurents, cafes, etc..) available in and around train stations in Europe. 

### Data description

Europe has 817 major and 56173 minor train stations. This is evident from data collected through github repository: https://github.com/trainline-eu/stations. The presented data was part of a kaggle dataset. The much needed details of venues was obtained through Foursquare API.
Original dataframe contains 12 columns and 56,990 rows. Due to limitions in Foursquare API requests, only major train stations in Europe were considered for this project. 

Some columns provide no valuable input to our project and are removed. When major train stations are considered, number of rows were down to 791.
After removing unnecessary columns, we will have columns with name of train stations, latitude and longitude values.

The latitude and longitude values provided in the dataframe are of train stations, but not of the city or region. This can be observed in below exaple of comparing "Menton" train station coordinates with "Menton" city coordinates



## Methodology

After clean dataset is available for major train stations in Europe, we will start by visualizing the data. Using folium library, a map of Europe is created with each station highlighted in purple color circle. From the map, one can realize that a significant number of stations are located in Switzerland. The data provides locations heavily centered in western europe. While train stations of capital cities of eastern europe were also listed in our dataset. One more observation is that major train stations are located is higher frequency around financial hubs (such as Greater London area, Frankfurt, Rhine valley, Paris-Reim, etc..)  
 

Create a Foursquare API GET request to gather information on venues, venue latitudes, venue longitudes and venue category based on coordinates of train stations. This information is collected in the form of a dataset and merged onto train stations dataframe. This new dataframe contains columns with data on Train station names, latitudes, longitudes, venue names, latitude, longitude of venues and venue category. The rows that contain venue category values as "Train station" were removed as first column implies the same details.



The values in "Venue category" column are one hot encoded with respective "Train station" value and a new dataframe is created with this values. With this dataframe, a new dataframe that lists most common venues at each major train station is created.


We are taking K-means clustering approach to create clusters of train stations that have more similarity. A total of seven clusters are created and this can be noted by added new column "Cluster_Label". This column implies the group to which the train station belongs. ## Results
Initial dataframe consisted of 791 major stations across Europe. This data was then merged with venue details obtained from Foursquare API. A total of 7 clusters were created.
First cluster (Highlighted by red circles) consists of 66 stations

Second cluster (Highlighted by purple circles) consists of 85 stations.

Third cluster (Highlighted by blue circles) contains 90 stations.

Fourth cluster (Highlighted by navy blue circles) consists of 470 stations.

Fifth cluster (Highlighted by light blue circles) lists 59 stations.

Sixth cluster (Highlighted by beige circles) has 16 stations.

Seventh cluster(Highlighted by orange circles) has 5 stations.


## Discussion

A number of train stations in First cluster are in southern europe. Most common venues in the first cluster are cafes, bars, Italian restaurents, ice cream shops, etc..This cluster has atleast one public building/service among common venues such as Zoo, exhibit, financial or legal service. Second cluster has stations that are surrounded by service based facilities such as super markets, gas stations and stores. In addition to these facilities, a couple of diners or restaurents were in most favoured venues. This cluster has majority of stations spread across France, Germany and Switzerland.

Entertainment venues are more common in third clusters. Facilities such as hotels, pools, bars, spas and nightclubs are around the stations that fall into third clusters. Apart from these, atleast one public building is available near the third cluster. Forth cluster covers over 50% of all stations in Europe. This cluster has a good combination of various restaurents, stores and public buildings. Every station in this cluster has 5 or 6 restaurents among the most common venues. Each restaurent has different cuisine. The remaining top venues consists of stores such as pharmacy, game/toy stores, markets along with service buildings.

Fifth cluster has stations that has a mix of venues and no definitive pattern. Sixth cluster has construction spaces, farms and service buildings as most common venues. Restaurents and stores are least favourable here. Seventh cluster has banks, Zoos, factories and fairs in thier vicinities.




## Conclusion

Majority of train stations in Italy and Croatia falls under red cluster that has cafes, bars, italian restaurents along with public service buildings. If investors are interested in these countries, they can invest in common venues of red cluster.

In western europe (excluding Italy and Croatia), Fourth cluster has most common venues and is safest best for investment.

If investor is interested in stations in Switzerland, a combination of fourth, first and second clusters are present. These three clusters have restaurents and entertainment venues as safe bet options.
